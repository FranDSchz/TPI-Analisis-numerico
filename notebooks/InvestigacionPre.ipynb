## Punto 1: Introducción a NumPy

**NumPy** es una biblioteca de Python que proporciona soporte para **arreglos y matrices multidimensionales**, junto con **funciones matemáticas de alto nivel**. Es ideal para realizar operaciones de cálculo numérico como la **multiplicación de matrices**, la **inversa de matrices** y otros cálculos algebraicos.


### Operaciones con Matrices en NumPy
* Multiplicación de Matrices
La multiplicación de matrices en NumPy se realiza con la función `np.dot()` o con el operador `@`. La multiplicación de matrices sigue la regla de producto de matrices: el número de columnas de la primera matriz debe ser igual al número de filas de la segunda.



```python
import numpy as np
# Definir matrices A y B
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
# Multiplicación de matrices
producto_matriz = np.dot(A, B)
# O alternativamente: producto_matriz = A @ B
print("Producto de A y B:\n", producto_matriz)
```

* Cálculo de la Inversa de una Matriz
La inversa de una matriz es una matriz que, cuando se multiplica por la original, da como resultado la matriz identidad. No todas las matrices tienen una inversa, sólo aquellas que son cuadradas y no singulares.


```python
# Matriz cuadrada
A = np.array([[1, 2], [3, 4]])
# Cálculo de la inversa
inversa_A = np.linalg.inv(A)
print("Inversa de A:\n", inversa_A)
```

* Cálculo de Valores Propios y Vectores Propios
Los valores propios (autovalores) y vectores propios (autovectores) son útiles para descomponer matrices y se usan en aplicaciones como la compresión de datos y la resolución de sistemas de ecuaciones diferenciales.


```python
# Definir matriz
A = np.array([[4, -2], [1, 1]])
# Calcular valores propios y vectores propios
valores_propios, vectores_propios = np.linalg.eig(A)
print("Valores propios de A:\n", valores_propios)
print("Vectores propios de A:\n", vectores_propios)
```

## Punto 2: Resolucion de sistemas de ecuaciones lineales con NumPy

Para resolver un sistema de ecuaciones lineales con NumPy, se utiliza la función `np.linalg.solve()`, que resuelve sistemas de la forma AX=BAX = BAX=B, donde:
* AAA es la matriz de coeficientes del sistema.
* XXX es el vector de incógnitas que queremos encontrar.
* BBB es el vector de términos independientes.

NumPy utiliza métodos de factorización para encontrar la solución de manera eficiente.

### Proceso para Resolver un Sistema de Ecuaciones Lineales con NumPy

1. Definir la Matriz de Coeficientes AAA: Representa los coeficientes de las incógnitas en cada ecuación. Asegúrate de que la matriz sea cuadrada (tenga el mismo número de filas y columnas) y que el sistema tenga solución.

2. Definir el Vector de Términos Independientes BBB: Este vector contiene los resultados de cada ecuación del sistema.

3. Usar `np.linalg.solve(A, B)`: Esta función toma la matriz AAA y el vector BBB y devuelve el vector de incógnitas XXX que satisface la ecuación AX=BAX = BAX=B.


### Explicación de la Salida

El código anterior imprimirá los valores de xxx y yyy que satisfacen el sistema de ecuaciones dado. En este caso, `np.linalg.solve(A, B)` encuentra automáticamente los valores de xxx y yyy utilizando métodos de álgebra lineal.

### Consideraciones Importantes 

* Matriz No Singular: La matriz AAA debe ser invertible (no singular) para que haya una solución única. Si AAA no es invertible, `np.linalg.solve()` generará un error indicando que el sistema no tiene solución única.

* Consistencia: Asegúrate de que las dimensiones de AAA y BBB sean compatibles. Por ejemplo, si AAA es de n×nn \times nn×n, entonces BBB debe ser un vector de dimensión nnn.

```python
#Ejemplo

# Matriz de coeficientes A
A = np.array([[2, 3], [1, -1]])
# Vector de terminos independientes B
B = np.array([8, 2])
# Resolver el sistema de ecuaciones
solucion = np.linalg.solve(A, B)
print("La solucion es: x =", solucion[0], ", y =", solucion[1])
```

## Punto 3: Introducción a SymPy

**SymPy** es una librería de Python diseñada para realizar cálculos matemáticos simbólicos, es decir, trabaja con expresiones y ecuaciones de forma algebraica en lugar de solo valores numéricos. Esto es especialmente útil en áreas de matemáticas donde es necesario manipular expresiones simbólicas, como álgebra, cálculo diferencial e integral, y ecuaciones diferenciales.

### Uso de SymPy en el Cálculo Diferencial

En el contexto del cálculo diferencial, SymPy permite:

1. **Resolver Ecuaciones Diferenciales Ordinarias (EDOs) y Ecuaciones Diferenciales Parciales (EDPs)** de manera simbólica, encontrando soluciones exactas.

2. **Derivación e Integración Simbólica**, que es útil para obtener la derivada o integral de una función sin necesidad de valores numéricos específicos.

3. **Series de Taylor** para aproximaciones de funciones.

4. **Análisis de estabilidad y puntos críticos** para sistemas de ecuaciones diferenciales, facilitando la clasificación de puntos de equilibrio.


### Resolución de Ecuaciones Diferenciales con SymPy

SymPy cuenta con la función `dsolve()` que permite resolver ecuaciones diferenciales de forma simbólica. Al ser simbólica, esta función no solo calcula una solución numérica para un valor específico, sino que encuentra una expresión general de la solución.

**Ejemplo Básico**

```python
from sympy import symbols, Function, Eq, dsolve
# Definir la variable independiente y la función dependiente
x = symbols('x')
y = Function('y')


# Definir la ecuación diferencial
edo = Eq(y(x).diff(x), y(x))
# Resolver la ecuación diferencial
solucion = dsolve(edo, y(x))
print(solucion)
```

### Ventajas de SymPy sobre Métodos Numéricos

1. **Soluciones Exactas**: SymPy encuentra soluciones exactas en forma de expresiones simbólicas, mientras que los métodos numéricos solo ofrecen aproximaciones en puntos específicos.
2. **Manipulación Algebraica**: Permite simplificar, factorizar o expandir expresiones, y trabajar con las soluciones de manera algebraica para análisis más profundos.
3. **Flexibilidad para Estudio Teórico**: Al ser simbólico, facilita el análisis teórico, como encontrar derivadas, integrales, o soluciones en función de parámetros.
4. **Identificación de Comportamientos Generales**: Con SymPy se pueden analizar tendencias y comportamientos de las soluciones sin necesidad de especificar un valor inicial o un punto en particular, como en estabilidad de puntos de equilibrio.

### Limitaciones de SymPy
Aun así, los métodos simbólicos como los de SymPy tienen sus limitaciones:
* No todos los sistemas de ecuaciones diferenciales tienen soluciones exactas, por lo que SymPy puede no encontrar una solución en algunos casos.

* Para problemas que requieren respuestas rápidas en aplicaciones prácticas (como modelos en tiempo real), los métodos numéricos pueden ser preferibles.


## Punto 4 : Introducción a SciPy en la Resolución de Ecuaciones Diferenciales

**SciPy** es una biblioteca de Python ampliamente utilizada para cálculos científicos y de ingeniería, especialmente útil para resolver ecuaciones diferenciales ordinarias (EDOs) mediante métodos numéricos. SciPy permite aproximar soluciones en intervalos específicos cuando no es posible hallar una solución exacta.

### Principales Funciones en SciPy para Resolver Ecuaciones Diferenciales

Las dos funciones más utilizadas en SciPy para resolver ecuaciones diferenciales ordinarias son `odeint` y `solve_ivp`.

#### 1. `odeint`
- Esta función es parte del módulo `scipy.integrate` y utiliza el solver `LSODA`, ideal para ecuaciones que alternan entre rigidez y no rigidez.
- `odeint` es sencilla y fácil de usar en problemas básicos o de rigidez mixta.
- La función toma como argumentos una función que define la ecuación diferencial, el estado inicial y un rango de tiempo.

##### Ejemplo de uso de `odeint`:

```python
from scipy.integrate import odeint
import numpy as np

# Definir la función del sistema
def modelo(y, t):
    dydt = -y
    return dydt

# Condición inicial
y0 = 5
# Intervalo de tiempo
t = np.linspace(0, 10, 100)

# Resolver la ecuación
solucion = odeint(modelo, y0, t)
print(solucion)
```

#### 2. `solve_ivp`
- `solve_ivp` es otra función en `scipy.integrate`, diseñada para ser más versátil y moderna.
- Permite utilizar métodos específicos como `RK45`, `RK23`, `BDF`, y otros, adaptándose a distintos tipos de problemas.
- Acepta una mayor cantidad de parámetros, como tolerancias y puntos de evaluación, y devuelve resultados más detallados.

##### Ejemplo de uso de `solve_ivp`:

```python
from scipy.integrate import solve_ivp

# Definir la función del sistema
def modelo(t, y):
    dydt = -y
    return dydt

# Intervalo de tiempo y condición inicial
tiempo = (0, 10)
y0 = [5]

# Resolver la ecuación con el método RK45
solucion = solve_ivp(modelo, tiempo, y0, method='RK45', t_eval=np.linspace(0, 10, 100))
print(solucion.y)
```

### ¿Cuándo Usar `odeint` o `solve_ivp`?

- **`odeint`**: Adecuada para problemas simples que requieren una configuración mínima. Si se necesita una solución rápida y no es necesario ajustar parámetros avanzados, esta función es una buena opción.
- **`solve_ivp`**: Recomendada para problemas complejos que requieren personalización. Ofrece control sobre tolerancias y métodos, ideal para análisis detallado en problemas rígidos y no rígidos.


## Punto 5: Representación de campos vectoriales con Matplotlib

Para representar y analizar campos vectoriales en dos dimensiones, se puede utilizar la biblioteca **Matplotlib** en Python, la cual ofrece una función especializada llamada `quiver` para graficar campos vectoriales. Un campo vectorial es un conjunto de vectores definidos en cada punto de una región del espacio, los cuales pueden representar diversas magnitudes como el **flujo de un fluido**, la **fuerza en un campo eléctrico o magnético**, o la **dirección del viento** en meteorología.


```python
import matplotlib.pyplot as plt 

#Para graficar un campo vectorial, es necesario definir una
#cuadrícula de puntos en el plano donde se visualizarán los 
# vectores. En este caso, los puntos x e y son generados mediante 
# la función np.linspace(-2, 2, 10), que crea 10 puntos equiespaciados
#  entre -2 y 2 para ambos ejes.

x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)

#Esta funcion convierte estos vectores x e y en dos matrices X e Y que 
# representan las coordenadas de cada punto en la cuadrícula.
#  Estas matrices serán la base para colocar cada vector del campo.
X, Y = np.meshgrid(x, y)

# Definir los componentes del campo vectorial
U = -Y
V = X

# Crear el gráfico del campo vectorial
plt.figure()

#Funcion quiver para graficar el campo vectorial
#X e Y: matrices de coordenadas para el punto de origen de cada vector.
#U y V: matrices que contienen las componentes del vector en x e y.
#angles='xy': especifica que los ángulos de los vectores se calculan en el plano xy.
#scale_units='xy': define las unidades de escala en el plano xy.
#scale=1: controla la longitud de las flechas (en este caso, se ha fijado sin escala adicional).
#color='r': asigna el color rojo a los vectores para mejorar la visibilidad.
plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1, color='r')

#Configurar los límites de los ejes y agregar etiquetas
plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Campo Vectorial')
plt.grid()
plt.show()
```

### **1. Resolución de ecuaciones diferenciales (SymPy vs. SciPy)**

> - *Toma una ecuación diferencial de la guía de trabajos prácticos y resuélvela usando **SymPy** y **SciPy** respectivamente.*
> - *Compara los enfoques simbólico y numérico.*
> - *¿En qué casos preferirías usar uno u otro método?*

Vamos a resolver la siguiente ecuacion diferencial usando SymPy para obtener la solucion simbolica y SciPy para calcular la solucion numerica.

$y'' + 9y = e^t, y(0) = 0 ; y'(0) = 0$


#### **Solución Simbólica con SymPy**
En SymPy, resolveremos esta EDO de segundo orden usando `dsolve`, que permite encontrar soluciones exactas en términos de expresiones matemáticas.
#### **Paso a Paso en SymPy**
1. Definimos las variables simbólicas **𝑦** y **𝑡**
2. Planteamos la ecuación diferencial.
3. Aplicamos `dsolve` para resolver la ecuación.
4. Usamos las condiciones iniciales para obtener una solución particular.

Aquí está el código para hacerlo:

```python
import sympy as sp

# Definimos la variable y la función desconocida
t = sp.symbols('t')
y = sp.Function('y')(t)

# Definimos la ecuación diferencial
ode = sp.Eq(y.diff(t, 2) + 9*y, sp.exp(t))

# Resolvemos la EDO simbólicamente
solucion_simb = sp.dsolve(ode, y)

# Aplicamos las condiciones iniciales y obtenemos la solución particular
constantes = sp.solve([solucion_simb.rhs.subs(t, 0), solucion_simb.rhs.diff(t).subs(t, 0)], 'C1', 'C2')
solucion_particular = solucion_simb.subs(constantes)

solucion_particular
```

Este código te dará una solución exacta en términos de funciones y exponentes. La solución exacta, de este modo, resulta útil para el análisis detallado de la solución, encontrar patrones o analizar el comportamiento a medida que $t \to \infty$

#### **Solución Numérica con SciPy**
Para una solución numérica, usamos `solve_ivp` de **SciPy**, un método de integración numérica. En este enfoque, convertimos la EDO de segundo orden en un sistema de EDOs de primer orden:

Definimos:
* $y_1 = y$
* $y_2 = y'$

La EDO se convierte en:

* $y_1' = y_2$
* $y_2' = e^t - 9y_1$

Entonces podemos resolver este sistema de ecuaciones.

##### **Paso a Paso en SciPy**
1. Definimos el sistema de ecuaciones como una función.
2. Usamos `solve_ivp` para resolver la EDO numéricamente, aplicando las condiciones iniciales.

Aquí está el código para hacerlo:

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Definimos la función para el sistema de ecuaciones
def sistema(t, Y):
    y1, y2 = Y
    dy1_dt = y2
    dy2_dt = np.exp(t) - 9*y1
    return [dy1_dt, dy2_dt]

# Condiciones iniciales
Y0 = [0, 0]  # y(0) = 0, y'(0) = 0

# Intervalo de tiempo para la solución numérica
t_span = (0, 5)
t_eval = np.linspace(0, 5, 100)

# Resolvemos la EDO numéricamente
solucion_num = solve_ivp(sistema, t_span, Y0, t_eval=t_eval)

# Graficamos la solución numérica
plt.plot(solucion_num.t, solucion_num.y[0], label="Solución numérica")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.legend()
plt.show()
```

Este código mostrará una gráfica de la solución numérica en el intervalo de tiempo seleccionado.

### **Comparación de los Enfoques**
| Aspecto            | SymPy (Simbólico)                               | SciPy (Numérico)                                           |
|--------------------|-------------------------------------------------|------------------------------------------------------------|
| Resultado          | Expresión exacta                                | Valores numéricos en puntos discretos                      |
| Precisión          | Solución exacta (si es posible)                 | Aproximación numérica                                      |
| Aplicación         | Útil para análisis teórico                      | Útil para simulaciones o problemas con parámetros no conocidos |
| Limitaciones       | Requiere que SymPy pueda resolver la EDO        | Puede ser lento en problemas de alta complejidad o intervalos largos |
| Ejemplos de Uso    | Estudio de comportamiento asintótico, estabilidad | Simulaciones en ingeniería, física y biología              |

### **¿Cuándo usar uno u otro?**
* **SymPy (Simbólico)**: Es ideal cuando necesitas una solución general que puedas manipular algebraicamente o para estudios teóricos, ya que proporciona una expresión completa y exacta.
* **SciPy (Numérico)**: Es preferible cuando te enfrentas a ecuaciones complejas donde no se puede encontrar una solución simbólica o cuando estás interesado en una simulación que requiere eficiencia y precisión en puntos específicos.

### **2. Transformada de Laplace con SymPy**

> * *Investiga cómo realizar la Transformada de Laplace utilizando SymPy.*
> * *Explica el uso de las funciones laplace_transform e inverse_laplace_transform.*
> * *¿En qué aplicaciones prácticas crees que esta herramienta es útil?*

La Transformada de Laplace es una herramienta matemática poderosa que convierte una función en el dominio del tiempo en una función en el dominio de la frecuencia, lo que facilita el análisis de sistemas dinámicos, como circuitos eléctricos y sistemas mecánicos. En **SymPy**, podemos realizar tanto la Transformada de Laplace como su inversa (para volver al dominio del tiempo) usando las funciones `laplace_transform` y `inverse_laplace_transform`.

#### **Transformada de Laplace en SymPy**

La función laplace_transform en SymPy tiene la siguiente sintaxis: `sympy.laplace_transform(función, variable, variable_transformada)`

* `función`: Es la función en el dominio del tiempo que queremos transformar
* `variable`: Es la variable de la función en el dominio del tiempo, generalmente **𝑡**
* `variable_transformada`: Es la variable del dominio de la frecuencia, comúnmente **𝑠**

Esta función devuelve tres elementos:

1. La función transformada.
2. La región de convergencia (donde la transformada es válida).
3. Una bandera que indica si la transformada fue calculada.

**Ejemplo de Transformada de Laplace**

Supongamos que queremos encontrar la Transformada de Laplace de la función $f(t) = e^{3t} \sin(2t)$

```python
import sympy as sp

# Definimos la variable del tiempo y de la frecuencia
t, s = sp.symbols('t s')
f_t = sp.exp(3 * t) * sp.sin(2 * t)

# Calculamos la Transformada de Laplace
F_s = sp.laplace_transform(f_t, t, s)
F_s
```

Este código nos devuelve la expresión de la transformada en términos de **𝑠**.

#### **Transformada de Laplace Inversa en SymPy**

La Transformada de Laplace Inversa nos permite regresar una función del dominio de la frecuencia al dominio del tiempo. La función `inverse_laplace_transform` tiene esta sintaxis: `sympy.inverse_laplace_transform(función_transformada, variable_transformada, variable)`

* `función_transformada`: Es la función en el dominio de la frecuencia que queremos invertir.
* `variable_transformada`: Es la variable del dominio de la frecuencia, generalmente **𝑠**.
* `variable`: Es la variable en el dominio del tiempo, comúnmente **𝑡**

**Ejemplo de Transformada de Laplace Inversa**

Supongamos que queremos encontrar la Transformada de Laplace Inversa de la función

$F(s) = \frac{s + 3}{s^2 + 4}$

```python
# Definimos la función en el dominio de la frecuencia
F_s = (s + 3) / (s**2 + 4)

# Calculamos la Transformada de Laplace Inversa
f_t = sp.inverse_laplace_transform(F_s, s, t)
f_t
```

Este código devolverá una expresión en términos de **𝑡**, correspondiente a la función en el dominio del tiempo.

#### **Aplicaciones Prácticas de la Transformada de Laplace**

La Transformada de Laplace se usa en una amplia gama de aplicaciones, especialmente en el análisis de sistemas lineales y de control. Algunas aplicaciones específicas incluyen:

1. **Análisis de Sistemas de Control**: La Transformada de Laplace facilita el análisis de sistemas dinámicos, como sistemas eléctricos, mecánicos y de control, permitiendo resolver ecuaciones diferenciales y estudiar la respuesta de sistemas a diferentes entradas.

2. **Circuitos Eléctricos**: En la teoría de circuitos, la Transformada de Laplace ayuda a analizar la respuesta de circuitos RLC (resistor, inductor, capacitor) al pasar al dominio de la frecuencia, simplificando la resolución de sistemas complejos.

3. **Ingeniería Mecánica y Vibraciones**: En mecánica, es útil para analizar sistemas vibratorios, como masas y resortes, ya que permite estudiar las respuestas a fuerzas externas, amortiguamiento y resonancia.

4. **Resolución de Ecuaciones Diferenciales**: En matemáticas aplicadas y física, la Transformada de Laplace permite resolver ecuaciones diferenciales con condiciones iniciales complejas, como las que surgen en el modelado de sistemas.

#### **Resumen**

La Transformada de Laplace y su inversa en SymPy permiten resolver problemas simbólicamente, proporcionando soluciones exactas en el dominio de la frecuencia. Esto es particularmente útil para analizar sistemas donde las ecuaciones diferenciales son complejas y el tiempo es una variable clave, lo que simplifica el estudio de la estabilidad y la respuesta de los sistemas en el tiempo y facilita el diseño y análisis de sistemas de control.

La librería **SymPy** facilita la creación y manipulación de matrices simbólicas, permitiendo el trabajo con entradas que no necesariamente son valores numéricos, sino expresiones o variables simbólicas. Esto es muy útil en álgebra lineal simbólica, donde las matrices pueden representar sistemas de ecuaciones, transformaciones lineales, y otros cálculos que requieren soluciones generales en términos de variables.

#### **Creación de Matrices Simbólicas en SymPy**

En **SymPy**, las matrices simbólicas se crean utilizando la clase Matrix. Para crear una matriz con variables simbólicas, primero necesitamos definir las variables, y luego podemos construir la matriz con ellas.

**Ejemplo básico de creación de matriz simbólica**

Aquí crearemos una matriz ___A___ de 2x2 con variables simbólicas ___a, b, c y d___:

```python
import sympy as sp

# Definimos las variables simbólicas
a, b, c, d = sp.symbols('a b c d')

# Creamos la matriz simbólica
A = sp.Matrix([[a, b], [c, d]])
A
```

#### **Manipulación de Matrices Simbólicas**

Una vez creada la matriz simbólica, SymPy permite realizar diversas operaciones algebraicas, tales como calcular la **determinante**, **inversa**, **transpuesta**, y resolver sistemas lineales usando matrices simbólicas.

**Ejemplo de cálculo de la determinante**

Para calcular el determinante de la matriz ___A___:

```python
det_A = A.det()
det_A
```

El resultado será una expresión en términos de ___a, b, c y d,___ si existe.

**jemplo de cálculo de la inversa**

Podemos calcular la inversa de la matriz ___A___, siempre y cuando su determinante no sea cero.

```python
inv_A = A.inv()
inv_A
```

Esto devuelve la inversa de ___A___ en términos de ___a, b, c y d,___ si existe.

**Ejemplo de multiplicación de matrices**

Supongamos que tenemos otra matriz simbólica ___B___:

```python
# Definimos otra matriz simbólica
e, f, g, h = sp.symbols('e f g h')
B = sp.Matrix([[e, f], [g, h]])

# Multiplicamos A por B
C = A * B
C
```

El resultado será la matriz ___C___ obtenida al multiplicar 
___A___ y ___B___.

**Ejemplo Completo: Sistema de Ecuaciones Lineales Simbólico**

Supongamos que tenemos un sistema de ecuaciones lineales simbólico:
$$
\begin{cases}
ax + by = e \\
cx + dy = f
\end{cases}
$$
Podemos expresar este sistema en forma de matriz y resolverlo simbólicamente.



```python
# Creamos el vector de incógnitas y el vector de términos independientes
x, y = sp.symbols('x y')
variables = sp.Matrix([x, y])
constants = sp.Matrix([e, f])

# Escribimos el sistema en forma matricial: A * variables = constants
solution = A.LUsolve(constants)  # Resuelve el sistema
solution
```

Este código resuelve el sistema en términos de las variables ___a, b, c, d, e___ y ___f___, devolviendo los valores de ___x___ y ___y___ en función de estas.

### **1. Derivadas simbólicas con SymPy**

#### **1. Derivada Simple**

Una derivada simple es la derivada de una función con respecto a una sola variable.

**Ejemplo de Derivada Simple**
Supongamos que tenemos la función $f(x) = x^3 + 2x^2 + x + 1$ y queremos calcular su derivada con respecto a $x$.

```python
import sympy as sp

# Definir la variable simbólica
x = sp.symbols('x')

# Definir la función
f = x**3 + 2*x**2 + x + 1

# Calcular la derivada con respecto a x
f_prime = sp.diff(f, x)
f_prime
```

#### **2. Derivada Parcial**

Una derivada parcial se aplica a funciones de varias variables y calcula la tasa de cambio con respecto a una de las variables, manteniendo las demás constantes.

**Ejemplo de Derivada Parcial**

Supongamos que tenemos la función $g(x,y) = x^2 y + 3xy^2 + y^3$ y queremos calcular su derivada con respecto a $x$ y $y$.

```python
# Definir las variables simbólicas
x, y = sp.symbols('x y')

# Definir la función
g = x**2 * y + 3 * x * y**2 + y**3
g
```

```python
# Derivada parcial con respecto a x
g_partial_x = sp.diff(g, x)
print(f"Derivada parcial respecto de x: ")
g_partial_x
```

```python
# Derivada parcial con respecto a y
g_partial_y = sp.diff(g, y)
print(f"Derivada parcial respecto de y: ")
g_partial_y
```

### **Resumen de los Comandos**

* **Derivada Simple**: `diff(f, x)` donde `f` es la función y `x` la variable de derivación.

### **2. Cálculo de límites simbólicos con SymPy**

> * *Explora cómo se calcula un límite simbólico usando SymPy.*
> * *Investiga la función limit y proporciona un ejemplo de cómo se calcularía el límite de una función cuando una variable tiende a un valor específico.*

En **SymPy**, la función `limit()` se utiliza para calcular el límite simbólico de una función cuando una variable se aproxima a un valor específico. Esta función permite calcular límites laterales (cuando se acerca desde la izquierda o desde la derecha) y es muy útil para trabajar con funciones en puntos de indeterminación.

#### **Sintaxis Básica de limit**

La sintaxis para la función `limit()` es:

`limit(función, variable, punto, dirección)`

* **`función`**: la expresión matemática de la cual queremos calcular el límite.
* **`variable`**: la variable que tiende a un valor específico.
* **`punto`**: el valor al que se aproxima la variable.
* `**dirección`**: (opcional): indica desde qué dirección se aproxima al valor. Puede ser `'+'` (desde la derecha) o `'-'` (desde la izquierda). Si no se especifica, SymPy calculará el límite general.

#### **Ejemplo 1: Límite General**

Supongamos que queremos calcular el límite de la función $f(x) = \frac{\sin(x)}{x}$ cuando $x$ tiende a 0.

```python
import sympy as sp

# Definir la variable y la función
x = sp.symbols('x')
f = sp.sin(x) / x

# Calcular el límite cuando x tiende a 0
limite_general = sp.limit(f, x, 0)
limite_general
```

#### **Ejemplo 2: Límite Lateral**

Ahora calcularemos el límite de la función $g(x) = \frac{1}{x}$ cuando $x$ tiende a 0 desde la derecha $(x \to 0^+)$ y desde la izquierda $(x \to 0^-)$.


```python
import sympy as sp

# Definir la variable y la función
x = sp.symbols('x')
g = 1 / x
```

```python
# Calcular el límite cuando x tiende a 0 desde la derecha
limite_derecha = sp.limit(g, x, 0, '+')
limite_derecha
```

```python
# Calcular el límite cuando x tiende a 0 desde la izquierda
limite_izquierda = sp.limit(g, x, 0, '-')
limite_izquierda
```

### **Resumen de los Comandos**
* **Límite General**: `limit(f, x, valor)` donde `f` es la función, `x` la variable, y valor el punto al que se aproxima.
* **Límite Lateral**: `limit(f, x, valor, '+')` o `limit(f, x, valor, '-')` para calcular el límite desde la derecha o la izquierda.

Estos cálculos permiten resolver límites simbólicos de forma exacta y manejar casos de indeterminación o discontinuidades con facilidad en **SymPy**.

El cálculo simbólico de límites con SymPy ofrece varias ventajas en comparación con el cálculo manual, especialmente cuando se trabaja con límites complicados o en situaciones donde el proceso analítico es laborioso. A continuación se detallan algunas de las principales ventajas y cómo SymPy facilita estos cálculos:

#### **Ventajas de SymPy en el Cálculo de Límites**

1. **Automatización y precisión**:
* SymPy automatiza el proceso de calcular límites, eliminando errores que pueden surgir en el cálculo manual, especialmente en pasos intermedios complejos.

2. **Capacidad de manejar casos de indeterminación**:
* En muchos límites, aparecen casos de indeterminación como $\frac{0}{0}, \infty - \infty, 0 \cdot \frac{\infty}{\infty}$. SymPy utiliza técnicas avanzadas, como el Teorema de L'Hôpital, para resolver estos casos automáticamente, lo cual simplifica el proceso.
* El cálculo manual de indeterminaciones requiere el uso de herramientas adicionales (derivadas, manipulación algebraica), lo cual SymPy puede hacer en un solo paso.

* SymPy permite calcular límites desde la izquierda o derecha de un punto usando una simple opción en la función `limit()`. Esto facilita el análisis de funciones en puntos de discontinuidad.
* En el cálculo manual, evaluar límites laterales en estos puntos puede ser especialmente difícil, especialmente si hay discontinuidades o asintotas en el punto de interés.

4. **Resolución de límites complejos y compuestos**:

* Para funciones complejas, que involucran múltiples términos, raíces, funciones trigonométricas, logaritmos, o exponenciales, el cálculo manual se vuelve largo y propenso a errores. SymPy puede simplificar estos cálculos, aplicando de manera automática propiedades y teoremas de límites.

5. **Rapidez en la obtención de resultados**:

* Al usar SymPy, el tiempo para calcular límites, incluso en casos complejos, se reduce significativamente en comparación con el cálculo manual. Esto es especialmente útil en aplicaciones que requieren resolver muchos límites en poco tiempo, como en cálculos de derivadas de funciones en puntos críticos o en análisis de estabilidad

#### **Cómo SymPy Maneja Límites Complicados**

SymPy está diseñado para identificar y aplicar automáticamente reglas específicas de límites. Algunas de las técnicas que utiliza incluyen:

* **Descomposición en Series de Taylor**: En algunos casos, SymPy expande funciones en series alrededor de un punto, lo cual es útil para evaluar límites de funciones cerca de puntos singulares.
* Uso de Teorema de L'Hôpital: SymPy aplica este teorema para resolver indeterminaciones de tipo $\frac{0}{0}, \frac{\infty}{\infty}$, aplicando automáticamente derivadas en el numerador y el denominador.
* **Simplificación Algorítmica**: SymPy intenta simplificar funciones mediante reescrituras algorítmicas, aplicando identidades trigonométricas, exponenciales y logarítmicas para expresar funciones en formas que permiten un cálculo de límite directo.

En comparación con el cálculo manual, SymPy no solo facilita el proceso, sino que también abre la puerta a analizar límites que de otro modo serían imposibles o poco prácticos de resolver manualmente. ## **1. Sistemas de ecuaciones diferenciales con SciPy**

> * *Investiga cómo SciPy puede utilizarse para resolver sistemas de ecuaciones diferenciales.*
> * *Busca ejemplos en los que se resuelvan sistemas acoplados de ecuaciones utilizando solve_ivp.*
> * *Explica brevemente los pasos clave.*

En **SciPy**, la función `solve_ivp` de su módulo `integrate` es útil para resolver sistemas de ecuaciones diferenciales ordinarias (ODEs) acopladas o independientes. Esta herramienta permite definir una función que representa el sistema de ecuaciones y calcular soluciones numéricas en un intervalo específico.

### **Pasos Clave para Resolver un Sistema de Ecuaciones Diferenciales con `solve_ivp`**

#### 1.**Definir el sistema de ecuaciones diferenciales**: Representamos el sistema de ecuaciones como una función en Python. Esta función recibe el tiempo $t$ y un vector de variables $y$ y devuelve el vector de sus derivadas.

#### 2.**Configurar las condiciones iniciales y el intervalo de tiempo**: Se especifica un vector de condiciones iniciales para cada variable del sistema en el tiempo inicial y el intervalo de tiempo sobre el que se quiere integrar.

#### 3.**Resolver el sistema usando `solve_ivp`**: Llamamos a `solve_ivp` proporcionando la función del sistema, el intervalo de tiempo, las condiciones iniciales y otros parámetros opcionales para el método de integración o para controlar la precisión de la solución.

#### 4.**Analizar los resultados**: `solve_ivp` devuelve un objeto con información sobre los tiempos y valores de las variables a lo largo de la integración, lo cual facilita el análisis y la visualización de los resultados.

### **Ejemplo: Resolver un Sistema Acoplado de Ecuaciones Diferenciales**

Ejemplo: Resolver un Sistema Acoplado de Ecuaciones Diferenciales


$$
\begin{cases}
\frac{dy_1}{dt} = 3y_1 - 4y_2 \\
\frac{dy_2}{dt} = 2y_1 - y_2
\end{cases}
$$


con condiciones iniciales $y_1(0) = 1$ y $y_2(0) = 0$ y un intervalo de tiempo de 0 a 5.

#### **Código en Python:**

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Definir el sistema de ecuaciones
def sistema(t, y):
    y1, y2 = y
    dy1_dt = 3 * y1 - 4 * y2
    dy2_dt = 2 * y1 - y2
    return [dy1_dt, dy2_dt]

# Intervalo de tiempo y condiciones iniciales
t_span = (0, 5)
y0 = [1, 0]

# Resolver el sistema de ecuaciones
sol = solve_ivp(sistema, t_span, y0, t_eval=np.linspace(0, 5, 100))

# Graficar los resultados
plt.plot(sol.t, sol.y[0], label='y1(t)')
plt.plot(sol.t, sol.y[1], label='y2(t)')
plt.xlabel('Tiempo t')
plt.ylabel('Soluciones y1 y y2')
plt.legend()
plt.title('Solución de un sistema de ODE acoplado con solve_ivp')
plt.show()
```

#### **Explicación del Código:**

1. **Función** `sistema`: Define las ecuaciones diferenciales donde `dy1_dt` y `dy2_dt` representan las derivadas de $y_1$ y $y_2$.
2. Parámetros de solve_ivp:
* `t_span` define el intervalo de integración $(0,5)$.
* $y(0)$ proporciona las condiciones iniciales.
* `t_eval` especifica los puntos de tiempo en los que deseamos evaluar la solución.
3. Visualización: Se grafican $y_1(t)$ y $y_2(t)$ en función del tiempo, permitiendo observar el comportamiento del sistema.

### **¿Cuándo Usar Este Método?**

`solve_ivp` es ideal para sistemas acoplados complejos donde no existe una solución analítica sencilla, como en modelos de dinámica poblacional, reacciones químicas y sistemas eléctricos. Este enfoque numérico permite encontrar soluciones aproximadas y explorar el comportamiento del sistema bajo diferentes condiciones iniciales y parámetros.

Usaremos **Mayavi** para crear una visualización 3D interactiva de un campo vectorial.

#### **Código en Python:**

```python
import numpy as np
from mayavi import mlab

# Crear la cuadrícula de puntos 3D
x, y, z = np.mgrid[-5:5:10j, -5:5:10j, -5:5:10j]

# Definir las componentes del campo vectorial
u = y  # Componente en X
v = -x  # Componente en Y
w = z  # Componente en Z

# Visualización del campo vectorial
mlab.quiver3d(x, y, z, u, v, w)

# Mostrar la ventana interactiva
mlab.show()
```

#### **Explicación del Código:**
1. **Datos en 3D:** `mgrid` genera una cuadrícula tridimensional de puntos.
2. **Campo Vectorial:** Los arreglos `u, v, y w` representan las componentes de cada vector en el campo.
3. **Gráfico 3D:** `mlab.quiver3d` crea el gráfico de flechas, representando el campo vectorial en cada punto.
4. **Interactividad:** Mayavi proporciona una ventana de visualización en la que puedes rotar y hacer zoom.

### **Comparación de Plotly y Mayavi**

| Aspecto            | Plotly                                                   | Mayavi                                      |
|--------------------|----------------------------------------------------------|---------------------------------------------|
| Facilidad de uso   | Intuitivo, fácil de integrar en Jupyter y sitios web.     | Más técnico, ideal para ciencia e ingeniería. |
| Interactividad     | Muy buena en navegadores, gráficos responsivos.           | Excelente, en aplicaciones de escritorio.   |
| Visualización 3D   | Gráficos claros, pero limitados en volumen de datos.      | Potente para grandes volúmenes de datos 3D. |
| Plataforma         | Ideal para web y Jupyter Notebooks.                       | Mejor en aplicaciones de escritorio.        |

Ambas bibliotecas son útiles dependiendo del entorno y las necesidades del proyecto. **Plotly** es adecuado para aplicaciones web e informes interactivos, mientras que **Mayavi** es más potente para visualización científica y simulaciones en entornos de escritorio.
