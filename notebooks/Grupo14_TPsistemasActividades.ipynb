Para resolver el sistema de ecuaciones diferenciales lineales usaremos el enfoque de autovalores y autovectores. Este sistema se puede representar en forma matricial como:

### **$X' = AX$**

donde


$$
\mathbf{X} = \begin{pmatrix} x \\ y \end{pmatrix} \quad \text{y} \quad \mathbf{A} = \begin{pmatrix} 2 & 2 \\ 1 & 3 \end{pmatrix}
$$

### **1. Buscamos los autovalores de $A$**
1. La solución general se encuentra determinando los autovalores y autovectores de la matriz $A$.
2. Calcularemos los autovalores resolviendo el polinomio característico de $A$:
#### $det(A−λI)=0$

Lo vamos a resolver usando  Python usando SymPy:

```python
import sympy as sp

# Definir variables y matriz
t = sp.symbols('t')
x, y = sp.symbols('x y')
A = sp.Matrix([[2, 2], [1, 3]])
```

```python
# Calcular autovalores
autovals = A.eigenvals()
autovals
```

```python
# Calcular autovectores
autovects = A.eigenvects()
autovects
```

Los autovalores del sistema son:
1. $\lambda_1 = 4$ con autovector asociado $\begin{pmatrix} 1 \\ 1 \end{pmatrix}$
2. $\lambda_2 = 1$ con autovector asociado $\begin{pmatrix} -2 \\ 1 \end{pmatrix}$


### **2. Formar la Solución General**
La solución general del sistema se obtiene combinando los términos de los autovalores y sus autovectores. Para cada autovalor $\lambda$ con autovector asociado $\mathbf{v}$, la solución es de la forma:

$\mathbf{X}(t) = c_1 e^{\lambda_1 t} \mathbf{v}_1 + c_2 e^{\lambda_2 t} \mathbf{v}_2$

Por lo tanto, la solución general será:

$\mathbf{X}(t) = c_1 e^{4t} \begin{pmatrix} 1 \\ 1 \end{pmatrix} + c_2 e^{t} \begin{pmatrix} -2 \\ 1 \end{pmatrix}$

La solución general del sistema es:

$\begin{pmatrix} x(t) \\ y(t) \end{pmatrix} = \begin{pmatrix} c_1 e^{4t} - 2c_2 e^{t} \\ c_1 e^{4t} + c_2 e^{t} \end{pmatrix}$

donde $c_1$ y $c_2$ son constantes arbitrarias que se determinan según las condiciones iniciales.

### **Graficar el campo vectorial asociado**
Para graficar el campo vectorial, necesitamos definir las funciones que representan las derivadas $\frac{dx}{dt}$ y $\frac{dy}{dt}$ del sistema de ecuaciones diferenciales.

Dado:
$$
\begin{cases}

## **Ejercicio 5.j**
### Paso 1: Resolver el sistema de ecuaciones diferenciales

Dado el sistema:

$\frac{dx}{dt} = x + y$

$\frac{dy}{dt} = -2x - y$

Podemos escribir el sistema en forma matricial:

$\frac{d}{dt} \begin{pmatrix} x \\ y \end{pmatrix} = \begin{pmatrix} 1 & 1 \\ -2 & -1 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix}$

Llamamos a la matriz \(A\):

$A = \begin{pmatrix} 1 & 1 \\ -2 & -1 \end{pmatrix}$

### **Calcular los autovalores y autovectores de $A$.**

Calculamos los autovalores resolviendo el polinomio característico:

$\det(A - \lambda I) = 0$

Para esta matriz, obtenemos los autovalores complejos:

```python
import sympy as sp

# Definir variables simbólicas
t = sp.symbols('t')
x, y = sp.symbols('x y')
z = sp.Matrix([x, y])

# Definir la matriz del sistema
A = sp.Matrix([[1, 1], [-2, -1]])

# Calcular autovalores y autovectores
autovalores = A.eigenvals()
autovectores = A.eigenvects()

# Imprimir autovalores y autovectores de forma más legible
print("Autovalores:")
sp.pprint(autovalores)

print("\nAutovectores:")
for eigen in autovectores:
    autovalor = eigen[0]
    multiplicidad = eigen[1]
    autovector = eigen[2][0]
    print(f"Autovalor: {autovalor}, Multiplicidad: {multiplicidad}")
    print("Autovector asociado:")
    sp.pprint(autovector)
    print("\n")
```

#### **Resultado esperado**
SymPy nos devuelve los autovalores complejos **($\lambda = ± i$)** y sus correspondientes autovectores.

Estos autovalores indican un comportamiento oscilatorio, por lo que sabemos que la solución general incluirá funciones seno y coseno.

### **Hallamos la solucion general**

Dado que tenemos autovalores complejos, planteamos la solución general en términos de seno y coseno usando SymPy.

```python
# Resolver el sistema diferencial simbólicamente
import sympy as sp

# Definir variables simbólicas
t = sp.symbols('t')
C1, C2 = sp.symbols('C1 C2')
func_x = sp.Function('x')(t)
func_y = sp.Function('y')(t)

# Definir las ecuaciones diferenciales
eq1 = sp.Eq(sp.diff(func_x, t), func_x + func_y)
eq2 = sp.Eq(sp.diff(func_y, t), -2 * func_x - func_y)

# Resolver el sistema
soluciones = sp.dsolve([eq1, eq2])

# Formatear y mostrar las soluciones
for sol in soluciones:
    display(sol)
```

Este código nos da la solución general del sistema en términos de $x(t)$ y $y(t)$

### **Resolvemos numéricamente con `solve_ivp`**

Usamos `solve_ivp` de SciPy para calcular la solución numérica y graficarla junto con el **campo vectorial**.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definir el sistema de ecuaciones
def sistema(t, z):
    x, y = z
    dxdt = x + y
    dydt = -2 * x - y
    return [dxdt, dydt]

# Crear una cuadrícula de puntos para el campo vectorial
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
U = X + Y
V = -2 * X - Y

# Graficar el campo vectorial
plt.figure(figsize=(8, 6))
plt.quiver(X, Y, U, V, color="blue")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Campo vectorial del sistema de ecuaciones')
plt.grid()
plt.show()

# Soluciones del sistema con condiciones iniciales
t_span = (0, 10)
z0 = [1, 0]  # Condiciones iniciales

sol = solve_ivp(sistema, t_span, z0, t_eval=np.linspace(0, 10, 200))

# Graficar las funciones solución
plt.figure(figsize=(10, 6))
plt.plot(sol.t, sol.y[0], label="x(t)")
plt.plot(sol.t, sol.y[1], label="y(t)")
plt.xlabel('t')
plt.ylabel('Valor de la función')
plt.title('Funciones solución x(t) y y(t)')
plt.legend()
plt.grid()
plt.show()
```

Esto significa que las trayectorias se mantienen en ciclos cerrados alrededor del origen, pero no se alejan ni se acercan, lo cual es característico de un **centro** en el plano fase.

> Resuelva los siguientes problemas de valores iniciales


$$
\mathbf{X}' = \begin{pmatrix}
2 & 4 \\
-1 & 6
\end{pmatrix} \mathbf{X}, \quad \mathbf{X}(0) = \begin{pmatrix}
-1 \\
6
\end{pmatrix}
$$

#### **Configurar e Inicializar las Variables y Matrices en SymPy**

Definimos la matriz $A$ del sistema y el vector de condiciones iniciales $X(0)$.

```python
import sympy as sp

# Definimos la variable de tiempo y el vector de incógnitas X(t)
t = sp.symbols('t')
x, y = sp.symbols('x y')
X = sp.Matrix([x, y])

# Matriz del sistema y vector de condiciones iniciales
A = sp.Matrix([[2, 4], [-1, 6]])
X0 = sp.Matrix([-1, 6])

# Mostramos la matriz A y la condición inicial
print("Matriz del sistema A:")
sp.pprint(A)

print("\nCondiciones iniciales X(0):")
sp.pprint(X0)
```

### **Calcular los Autovalores y Autovectores**

Ahora, calculamos los autovalores y autovectores de $A$, que son esenciales para resolver el sistema de ecuaciones diferenciales.

```python
# Cálculo de autovalores y autovectores
autovalores = A.eigenvals()
autovectores = A.eigenvects()

# Verificamos el resultado de autovalores y autovectores
print("\nAutovalores:")
sp.pprint(autovalores)

print("\nAutovectores:")
for autovalor, multiplicidad, autovector in autovectores:
    print(f"Autovalor: {autovalor}, Multiplicidad: {multiplicidad}")
    print("Autovector asociado:")
    sp.pprint(autovector[0])
    print("\n")
```

### **Construir la Solución General del Sistema**

Para construir la solución general, usamos la combinación lineal de los autovectores escalados por exponenciales de sus autovalores multiplicados por $t$.

```python
# Definimos las constantes de integración
C1, C2 = sp.symbols('C1 C2')

# Creamos la solución general con manejo para autovalores complejos
solucion_general = sp.Matrix([0, 0])
for i, (autovalor, multiplicidad, autovector) in enumerate(autovectores):
    # Combinamos autovectores con exponenciales de autovalores
    solucion_general += sp.symbols(f'C{i+1}') * autovector[0] * sp.exp(autovalor * t)

# Mostramos la solución general
print("Solución general del sistema:")
display(solucion_general)
```

### **Resolvemos las Constantes de Integración usando la Condición Inicial**

Sustituimos $t = 0$ en la solución general y usamos la condición inicial para resolver $C1$ y $C2$.

```python
# Igualamos la solución en t = 0 con el vector de condiciones iniciales
ecuacion_inicial = sp.Eq(solucion_general.subs(t, 0), X0)

# Resolvemos para C1 y C2
constantes = sp.solve(ecuacion_inicial, (C1, C2))

# Solución particular con las constantes específicas
solucion_particular = solucion_general.subs(constantes)

# Mostramos la solución particular
print("Solución particular del sistema:")
display(solucion_particular)
```

## **2: Sistemas no homogéneos con variación de parámetros**
___

Dado el sistema:

$$
\frac{d}{dt} \begin{pmatrix} x_1 \\ x_2 \end{pmatrix} = \begin{pmatrix} 1 & 8 \\ 1 & -1 \end{pmatrix} \begin{pmatrix} x_1 \\ x_2 \end{pmatrix} + \begin{pmatrix} e^{-t} \\ t e^t \end{pmatrix}
$$



```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Definimos la matriz A y el vector de términos independientes B(t)
A = sp.Matrix([[1, 8], [1, -1]])
B = sp.Matrix([sp.exp(-sp.Symbol('t')), sp.Symbol('t') * sp.exp(sp.Symbol('t'))])

# Paso 1: Resolver el sistema homogéneo asociado (X' = A * X)
# Calculamos los valores y vectores propios de A
eigvals = A.eigenvals()
eigvecs = A.eigenvects()

print("Valores propios de A:", eigvals)
print("Vectores propios de A:", eigvecs)

# Paso 2: Construir la matriz fundamental
t = sp.symbols('t')
solutions_homogeneous = []
for eigen in eigvecs:
    lam = eigen[0]  # valor propio
    vec = eigen[2][0]  # vector propio correspondiente
    sol = sp.exp(lam * t) * vec  # solución del sistema homogéneo
    solutions_homogeneous.append(sol)

# Creamos la matriz fundamental Φ(t)
Phi = sp.Matrix.hstack(*solutions_homogeneous)
print("Matriz fundamental Phi(t):")
display(Phi)

# Paso 3: Método de variación de parámetros
# Encontrar Phi_inv y Phi_inv * B
Phi_inv = Phi.inv()
u_prime = Phi_inv * B
u = sp.integrate(u_prime, t)  # Integrar para obtener u(t)
X_particular = Phi * u  # Solución particular
print("Solución particular Xp(t):")
display(X_particular)

# Paso 4: Solución general
X_general = Phi * sp.Matrix(sp.symbols('c1 c2')) + X_particular
print("Solución general X(t):")
display(X_general)

# Asignamos valores específicos a las constantes c1 y c2
X_general_numeric = X_general.subs({'c1': 1, 'c2': 1})

# Para visualizar los resultados, convertimos las expresiones a funciones numéricas y graficamos.
t_vals = np.linspace(0, 5, 100)
x1_func = sp.lambdify(t, X_general_numeric[0], 'numpy')
x2_func = sp.lambdify(t, X_general_numeric[1], 'numpy')

x1_vals = x1_func(t_vals)
x2_vals = x2_func(t_vals)

plt.figure(figsize=(10, 5))
plt.plot(t_vals, x1_vals, label="x1(t)")
plt.plot(t_vals, x2_vals, label="x2(t)")
plt.xlabel("t")
plt.ylabel("Solución")
plt.legend()
plt.title("Solución del sistema no homogéneo")
plt.grid()
plt.show()
```

Dado el siguiente sistema de ecuaciones diferenciales 


```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Definimos las variables simbólicas
t, s = sp.symbols('t s')
x = sp.Function('x')
y = sp.Function('y')

# Definimos la matriz del sistema y el vector no homogéneo
A = sp.Matrix([[1, -1], [2, 4]])
B = sp.Matrix([sp.exp(2 * t), 1])

# Condiciones iniciales
ics = {x(0): 1, y(0): 2}
X0 = sp.Matrix([ics[x(0)], ics[y(0)]])

# Transformada de Laplace del vector B
B_s = sp.laplace_transform(B, t, s, noconds=True)

# Definimos la matriz identidad
I = sp.eye(2)

# Planteamos el sistema en el dominio de Laplace: (sI - A) * X_s = X0 + B_s
lhs = s * I - A  # Matriz (sI - A)
rhs = X0 + B_s   # Vector lado derecho

# Resolvemos para X_s
X_s = lhs.inv() * rhs

# Extraemos las soluciones para X(s) y Y(s)
Xs_sol = X_s[0]
Ys_sol = X_s[1]

# Calculamos la transformada inversa de Laplace para obtener x(t) y y(t)
x_sol = sp.inverse_laplace_transform(Xs_sol, s, t)
y_sol = sp.inverse_laplace_transform(Ys_sol, s, t)

# Mostramos las soluciones
print("Solución para x(t):")
display(x_sol)
print("Solución para y(t):")
display(y_sol)

# Convertimos las soluciones a funciones numéricas
x_func = sp.lambdify(t, x_sol, modules=['numpy'])
y_func = sp.lambdify(t, y_sol, modules=['numpy'])

# Evaluamos las soluciones en un rango de tiempo para graficarlas
t_vals = np.linspace(0, 5, 100)
x_vals = x_func(t_vals)
y_vals = y_func(t_vals)

# Graficamos las soluciones
plt.figure(figsize=(10, 5))
plt.plot(t_vals, x_vals, label="x(t)")
plt.plot(t_vals, y_vals, label="y(t)")
plt.xlabel("t")
plt.ylabel("Solución")
plt.legend()
plt.title("Solución del sistema no homogéneo usando Transformada de Laplace")
plt.grid()
plt.show()


```

## **Consigna**
Dado el siguiente sistema de ecuaciones:
$$
\begin{pmatrix}
x_1'(t) \\
x_2'(t)
\end{pmatrix}
=
\begin{pmatrix}
0 & 1 \\
-1 & -k
\end{pmatrix}
\begin{pmatrix}
x_1(t) \\
x_2(t)
\end{pmatrix}
$$

Desarrolle el código necesario para analizar la estabilidad para los siguientes valores de \( k \):

a. $ k > 2 $

b. $ k = 2 $

c. $ 0 < k < 2 $

d. $ k = 0 $

e. $-2 < k < 0$

f. $k = -2 $

g. $k < -2$

Para analizar la estabilidad de este sistema de ecuaciones diferenciales, necesitamos estudiar los autovalores de la matriz del sistema para diferentes valores de k. La estabilidad del sistema dependerá de la naturaleza (parte real y signo) de los autovalores que obtenemos en cada caso.

### **Definir el sistema en Python**
Utilizaremos SymPy para definir la matriz del sistema, encontrar los autovalores en términos de k, y analizar su comportamiento

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

# Definimos las variables simbólicas
t = sp.symbols('t')
x1, x2 = sp.symbols('x1 x2')

# Definimos la matriz A en función de k
k = sp.symbols('k')
A = sp.Matrix([[0, 1], [-1, -k]])

# Definimos el vector de incógnitas
X = sp.Matrix([x1, x2])

# Definimos el sistema de ecuaciones
sistema = sp.Eq(X.diff(t), A * X)
```

### **Calculamos los autovalores de $A$**

```python
# Calcular los autovalores y autovectores de A
autovalores = A.eigenvals()
autovectores = A.eigenvects()

# Mostrar autovalores y autovectores
print("\nAutovalores de A:")
display(autovalores)

print("\nAutovectores de A:")
for autovalor, multiplicidad, autovector in autovectores:
    print(f"Autovalor: {autovalor}, Multiplicidad: {multiplicidad}")
    print("Autovector asociado:")
    display(autovector[0])
    print("\n")
```

### **Clasificar la estabilidad para diferentes valores de $k$**

Ahora que tenemos los autovalores en términos de $k$, analizaremos cada caso. Como guía, recordemos que:

* Si la parte real de todos los autovalores es negativa, el sistema es **estable** (las soluciones convergen al origen).
* Si algún autovalor tiene parte real positiva, el sistema es **inestable** (las soluciones divergen).
* Si la parte real es cero, el sistema puede ser **marginalmente estable o inestable**, dependiendo del tipo de autovalores.

```python
import sympy as sp

# Definir la función para analizar la estabilidad
def analizar_estabilidad(k):
    A = sp.Matrix([[0, 1], [-1, -k]])
    autovalores = A.eigenvals()
    
    print(f"\nAnálisis de estabilidad para k = {k}:")
    for autovalor in autovalores:
        # Evaluar la parte real del autovalor
        parte_real = sp.re(autovalor).evalf()
        if parte_real > 0:
            print(f"El autovalor {autovalor} tiene una parte real positiva ({parte_real}), el sistema es inestable.")
        elif parte_real < 0:
            print(f"El autovalor {autovalor} tiene una parte real negativa ({parte_real}), el sistema es estable.")
        else:
            print(f"El autovalor {autovalor} tiene una parte real cero ({parte_real}), el sistema es marginalmente estable.")

# Análisis de estabilidad para diferentes valores de k
print("a. k > 2")
for k in [3, 4, 5]:  # Añade otros valores si lo consideras necesario
    analizar_estabilidad(k)

print("\nb. k = 2")
analizar_estabilidad(2)

print("\nc. 0 < k < 2")
for k in [0.5, 1, 1.5]:  # Añade otros valores si lo consideras necesario
    analizar_estabilidad(k)

print("\nd. k = 0")
analizar_estabilidad(0)

print("\ne. -2 < k < 0")
for k in [-1.5, -1, -0.5]:  # Añade otros valores si lo consideras necesario
    analizar_estabilidad(k)

print("\nf. k = -2")
analizar_estabilidad(-2)

print("\ng. k < -2")
for k in [-3, -4, -5]:  # Añade otros valores si lo consideras necesario
    analizar_estabilidad(k)
```

### **Conclusion:**

Gracias a los resultados obtenidos con el algoritmo desarrollado podemos determinar la estabilidad para el sistema de ecuaciones en los diferentes intervalos:

#### **a. $k > 2$**

$k^2 - 4$ será positivo.

Los autovalores serán reales y distintos.

Ambos autovalores tendrán partes reales negativas ($\lambda_1, \lambda_2 < 0$).

**Estable**.

#### **b. $k = 2$**

$k^2 - 4 = 0$.

Los autovalores serán reales y iguales.

$\lambda = -1$.

**Estable**.

#### **c. $0 < k < 2$**

$k^2 - 4$ será negativo.

Los autovalores serán complejos conjugados con partes reales negativas.

$\text{Re}(\lambda_1, \lambda_2) = -\frac{k}{2}$.

**Estable**.

#### **d. $k = 0$**

$k^2 - 4 = -4$.

Los autovalores serán puramente imaginarios.

$\lambda_1, \lambda_2 = \pm i$.

**Marginalmente estable** (debido a la naturaleza oscilatoria sin amortiguación).

#### **e. $-2 < k < 0$**

$k^2 - 4$ será negativo.

Los autovalores serán complejos conjugados con partes reales positivas.

$\text{Re}(\lambda_1, \lambda_2) = -\frac{k}{2}$.

**Inestable**.

#### **f. $k = -2$**

$k^2 - 4 = 0$.

Los autovalores serán reales y repetidos.

$\lambda = 1$.

**Inestable**.

#### **g. $k < -2$**

$k^2 - 4$ será positivo.

Los autovalores serán reales y distintos.

Ambos autovalores tendrán partes reales positivas.

**Inestable**.

## **5: Funcion de Lyapunov para sistemas lineales**
___

Dado el siguiente sistema: 


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eig
from matplotlib import cm

# Definir la matriz del sistema
A = np.array([[0, 1], [-4, -5]])

# Calcular los autovalores del sistema
eigenvalues, _ = eig(A)
print("Autovalores del sistema:", eigenvalues)

# Definir una matriz P positiva definida para la función de Lyapunov
# En este caso, usaremos P = I (matriz identidad) para simplificar
P = np.eye(2)

# Definir la función de Lyapunov V(x) = x^T P x
def V(x):
    return np.dot(x.T, np.dot(P, x))

# Definir el campo vectorial del sistema
def system_dynamics(x, y):
    X = np.array([x, y])
    X_dot = A @ X
    return X_dot[0], X_dot[1]

# Crear una malla de puntos para evaluar y graficar
x_vals = np.linspace(-10, 10, 100)
y_vals = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = np.array([V(np.array([x, y])) for x, y in zip(X.ravel(), Y.ravel())])
Z = Z.reshape(X.shape)

# Graficar las curvas de nivel de la función de Lyapunov
plt.figure(figsize=(10, 8))
plt.contourf(X, Y, Z, levels=50, cmap=cm.viridis)
plt.colorbar(label='V(x) = x^T P x')

# Graficar el campo vectorial
U, V = np.vectorize(system_dynamics)(X, Y)
plt.quiver(X, Y, U, V, color="white")

# Configurar la gráfica
plt.title("Análisis de estabilidad con función de Lyapunov")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.show()
```

## Conclusiones del Análisis de Estabilidad con Función de Lyapunov

A partir del gráfico generado, se pueden extraer las siguientes conclusiones sobre la estabilidad del sistema:

### 1. Curvas de Nivel de la Función de Lyapunov
- Las curvas de nivel de la función \( V(x) = x^T P x \) muestran una disminución hacia el centro (el origen), indicando que la función de Lyapunov decrece conforme el sistema se acerca al punto de equilibrio.
- En el gráfico, los valores de \( V(x) \) son mayores en los bordes y disminuyen hacia el origen, lo que sugiere que la "energía" del sistema es mínima en el origen. Esto respalda la **estabilidad asintótica** del sistema.

### 2. Campo Vectorial
- Las flechas representan el campo vectorial del sistema. La dirección de estas flechas muestra que el sistema tiende a moverse hacia el origen, indicando una **atracción hacia el punto de equilibrio**.
- La disposición en espiral de las flechas sugiere un comportamiento de oscilación amortiguada, lo que es coherente con autovalores complejos con parte real negativa.

### 3. Estabilidad Asintótica
- Dado que el campo vectorial y las curvas de nivel de \( V(x) \) convergen hacia el origen y que la función de Lyapunov es positiva definida (es decir, \( V(x) > 0 \) para cualquier \( x \neq 0 \)), se concluye que el sistema es **estable asintóticamente**.
- Esto significa que cualquier perturbación inicial en el sistema se disipará con el tiempo, y el sistema regresará a su estado de equilibrio en el origen.

### 4. Interpretación Física
- Este comportamiento es característico de un **oscilador amortiguado**, en el que cualquier perturbación inicial se reduce hasta que el sistema permanece en reposo en el equilibrio. Esta es una propiedad deseable en sistemas que necesitan autorregularse y mantenerse estables frente a perturbaciones.

### Conclusión General
El gráfico confirma que el sistema es **estable asintóticamente**. La función de Lyapunov decrece hacia el origen, y el campo vectorial muestra que cualquier trayectoria en el espacio de estados tiende a moverse hacia el equilibrio en el origen, generalmente con un movimiento oscilatorio amortiguado. Esto implica que el sistema puede absorber perturbaciones iniciales y estabilizarse automáticamente.
