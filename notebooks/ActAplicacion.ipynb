> *Determine la soluci√≥n general del sistema de ecuaciones (Autovalores distintos)*



Para resolver el sistema de ecuaciones diferenciales lineales usaremos el enfoque de autovalores y autovectores. Este sistema se puede representar en forma matricial como:

### **$X' = AX$**

donde


$$
\mathbf{X} = \begin{pmatrix} x \\ y \end{pmatrix} \quad \text{y} \quad \mathbf{A} = \begin{pmatrix} 2 & 2 \\ 1 & 3 \end{pmatrix}
$$

### **1. Buscamos los autovalores de $A$**
1. La soluci√≥n general se encuentra determinando los autovalores y autovectores de la matriz $A$.
2. Calcularemos los autovalores resolviendo el polinomio caracter√≠stico de $A$:
#### $det(A‚àíŒªI)=0$

Lo vamos a resolver usando  Python usando SymPy:

```python
import sympy as sp

# Definir variables y matriz
t = sp.symbols('t')
x, y = sp.symbols('x y')
A = sp.Matrix([[2, 2], [1, 3]])
```

```python
# Calcular autovalores
autovals = A.eigenvals()
autovals
```

```python
# Calcular autovectores
autovects = A.eigenvects()
autovects
```

Los autovalores del sistema son:
1. $\lambda_1 = 4$ con autovector asociado $\begin{pmatrix} 1 \\ 1 \end{pmatrix}$
2. $\lambda_2 = 1$ con autovector asociado $\begin{pmatrix} -2 \\ 1 \end{pmatrix}$


### **2. Formar la Soluci√≥n General**
La soluci√≥n general del sistema se obtiene combinando los t√©rminos de los autovalores y sus autovectores. Para cada autovalor $\lambda$ con autovector asociado $\mathbf{v}$, la soluci√≥n es de la forma:

$\mathbf{X}(t) = c_1 e^{\lambda_1 t} \mathbf{v}_1 + c_2 e^{\lambda_2 t} \mathbf{v}_2$

Por lo tanto, la soluci√≥n general ser√°:

$\mathbf{X}(t) = c_1 e^{4t} \begin{pmatrix} 1 \\ 1 \end{pmatrix} + c_2 e^{t} \begin{pmatrix} -2 \\ 1 \end{pmatrix}$

La soluci√≥n general del sistema es:

$\begin{pmatrix} x(t) \\ y(t) \end{pmatrix} = \begin{pmatrix} c_1 e^{4t} - 2c_2 e^{t} \\ c_1 e^{4t} + c_2 e^{t} \end{pmatrix}$

donde $c_1$ y $c_2$ son constantes arbitrarias que se determinan seg√∫n las condiciones iniciales.

### **Graficar el campo vectorial asociado**
Para graficar el campo vectorial, necesitamos definir las funciones que representan las derivadas $\frac{dx}{dt}$ y $\frac{dy}{dt}$ del sistema de ecuaciones diferenciales.

Dado:
$$
\begin{cases}
\frac{dx}{dt} = 2x + 2y \\
\frac{dy}{dt} = x + 3y
\end{cases}
$$

### **Implementacion en python**

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una cuadr√≠cula de puntos en el plano (x, y)
x = np.linspace(-10, 10, 20)
y = np.linspace(-10, 10, 20)
X, Y = np.meshgrid(x, y)

# Definir el campo vectorial
dx = 2*X + 2*Y
dy = X + 3*Y

# Graficar el campo vectorial
plt.figure(figsize=(8, 6))
plt.quiver(X, Y, dx, dy, color='b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Campo vectorial del sistema de ecuaciones')
plt.grid()
plt.show()
```

### **Graficar las funciones soluci√≥n**
Para graficar las funciones soluci√≥n, definimos $x(t)$ y $y(t)$ en funci√≥n de t seg√∫n la soluci√≥n general obtenida:

$x(t) = c_1 e^{4t} - 2c_2 e^{t}$

$y(t) = c_1 e^{4t} + c_2 e^{t}$

Podemos elegir valores espec√≠ficos para $c_1$ y $c_2$ para ver c√≥mo evoluciona la soluci√≥n en el tiempo.

```python
# Definir valores de c1 y c2
c1 = 1
c2 = 1

# Crear un rango de valores de tiempo t
t = np.linspace(0, 2, 200)

# Calcular x(t) y y(t) con los valores de c1 y c2 elegidos
x_t = c1 * np.exp(4 * t) - 2 * c2 * np.exp(t)
y_t = c1 * np.exp(4 * t) + c2 * np.exp(t)

# Graficar x(t) y y(t)
plt.figure(figsize=(10, 6))
plt.plot(t, x_t, label='x(t)')
plt.plot(t, y_t, label='y(t)')
plt.xlabel('t')
plt.ylabel('Valor de la funci√≥n')
plt.title('Funciones soluci√≥n x(t) y y(t)')
plt.legend()
plt.grid()
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Definimos las variables simb√≥licas
t = sp.symbols('t')  # Variable de tiempo
x, y = sp.symbols('x y')  # Variables dependientes
X = sp.Matrix([x, y])  # Vector de variables
A = sp.Matrix([[-6, 5], [-5, 4]])  # Matriz del sistema

# ---
# ### 1. Soluci√≥n general del sistema
# Para encontrar la soluci√≥n general, calculamos los autovalores y autovectores de la matriz \( A \).
# ---
# Calculamos autovalores y autovectores
eigenvalues = A.eigenvals()
eigenvaluesAux= A.eigenvals()
eigenvectors = A.eigenvects()

# Mostramos los autovalores y autovectores
print("Autovalores:", eigenvalues)
print("Autovectores:")
for val, mult, vects in eigenvectors:
    print(f"Autovalor: {val}, multiplicidad: {mult}, autovector: {vects}")
```

---

Con los autovalores y autovectores, podemos escribir la soluci√≥n general:

$$
X(t) = C_1 e^{\lambda t} \vec{v} + C_2 t e^{\lambda t} \vec{v}
$$

donde \( \lambda \) es el autovalor repetido y \( \vec{v} \) es el autovector correspondiente.

---


```python
# Definimos las constantes de integraci√≥n
C1, C2 = sp.symbols('C1 C2')

# Ejemplo de valores para eigenvectors y eigenvalues
eigenvectors = [(1, 1, [sp.Matrix([1, 0]), sp.Matrix([0, 1])])]
eigenvalues = {sp.symbols('lambda'): 1}

# Verificamos que eigenvalues no est√© vac√≠o antes de usar popitem()
if eigenvalues:
    eigenvalue = eigenvalues.popitem()[0]
else:
    raise ValueError("El diccionario de eigenvalues est√° vac√≠o")

# Construimos la soluci√≥n general utilizando los autovectores y el autovalor repetido
sol = C1 * eigenvectors[0][2][0] * sp.exp(eigenvalue * t) + \
      C2 * t * eigenvectors[0][2][0] * sp.exp(eigenvalue * t)

# Descomponemos la soluci√≥n en x(t) y y(t)
sol_x = sol[0].subs({x: sp.Function('x')(t), y: sp.Function('y')(t)})
sol_y = sol[1].subs({x: sp.Function('x')(t), y: sp.Function('y')(t)})

print("Soluci√≥n general:")
sp.pprint(sol_x)
sp.pprint(sol_y)
```

## **Ejercicio 5.j**
### Paso 1: Resolver el sistema de ecuaciones diferenciales

Dado el sistema:

$\frac{dx}{dt} = x + y$

$\frac{dy}{dt} = -2x - y$

Podemos escribir el sistema en forma matricial:

$\frac{d}{dt} \begin{pmatrix} x \\ y \end{pmatrix} = \begin{pmatrix} 1 & 1 \\ -2 & -1 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix}$

Llamamos a la matriz \(A\):

$A = \begin{pmatrix} 1 & 1 \\ -2 & -1 \end{pmatrix}$

### **Calcular los autovalores y autovectores de $A$.**

Calculamos los autovalores resolviendo el polinomio caracter√≠stico:

$\det(A - \lambda I) = 0$

Para esta matriz, obtenemos los autovalores complejos:

```python
import sympy as sp

# Definir variables simb√≥licas
t = sp.symbols('t')
x, y = sp.symbols('x y')
z = sp.Matrix([x, y])

# Definir la matriz del sistema
A = sp.Matrix([[1, 1], [-2, -1]])

# Calcular autovalores y autovectores
autovalores = A.eigenvals()
autovectores = A.eigenvects()

# Imprimir autovalores y autovectores de forma m√°s legible
print("Autovalores:")
sp.pprint(autovalores)

print("\nAutovectores:")
for eigen in autovectores:
    autovalor = eigen[0]
    multiplicidad = eigen[1]
    autovector = eigen[2][0]
    print(f"Autovalor: {autovalor}, Multiplicidad: {multiplicidad}")
    print("Autovector asociado:")
    sp.pprint(autovector)
    print("\n")
```

#### **Resultado esperado**
SymPy nos devuelve los autovalores complejos **($\lambda = ¬± i$)** y sus correspondientes autovectores.

Estos autovalores indican un comportamiento oscilatorio, por lo que sabemos que la soluci√≥n general incluir√° funciones seno y coseno.

### **Hallamos la solucion general**

Dado que tenemos autovalores complejos, planteamos la soluci√≥n general en t√©rminos de seno y coseno usando SymPy.

```python
# Resolver el sistema diferencial simb√≥licamente
import sympy as sp

# Definir variables simb√≥licas
t = sp.symbols('t')
C1, C2 = sp.symbols('C1 C2')
func_x = sp.Function('x')(t)
func_y = sp.Function('y')(t)

# Definir las ecuaciones diferenciales
eq1 = sp.Eq(sp.diff(func_x, t), func_x + func_y)
eq2 = sp.Eq(sp.diff(func_y, t), -2 * func_x - func_y)

# Resolver el sistema
soluciones = sp.dsolve([eq1, eq2])

# Formatear y mostrar las soluciones
for sol in soluciones:
    display(sol)
```

Este c√≥digo nos da la soluci√≥n general del sistema en t√©rminos de $x(t)$ y $y(t)$

### **Resolvemos num√©ricamente con `solve_ivp`**

Usamos `solve_ivp` de SciPy para calcular la soluci√≥n num√©rica y graficarla junto con el **campo vectorial**.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definir el sistema de ecuaciones
def sistema(t, z):
    x, y = z
    dxdt = x + y
    dydt = -2 * x - y
    return [dxdt, dydt]

# Crear una cuadr√≠cula de puntos para el campo vectorial
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
U = X + Y
V = -2 * X - Y

# Graficar el campo vectorial
plt.figure(figsize=(8, 6))
plt.quiver(X, Y, U, V, color="blue")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Campo vectorial del sistema de ecuaciones')
plt.grid()
plt.show()

# Soluciones del sistema con condiciones iniciales
t_span = (0, 10)
z0 = [1, 0]  # Condiciones iniciales

sol = solve_ivp(sistema, t_span, z0, t_eval=np.linspace(0, 10, 200))

# Graficar las funciones soluci√≥n
plt.figure(figsize=(10, 6))
plt.plot(sol.t, sol.y[0], label="x(t)")
plt.plot(sol.t, sol.y[1], label="y(t)")
plt.xlabel('t')
plt.ylabel('Valor de la funci√≥n')
plt.title('Funciones soluci√≥n x(t) y y(t)')
plt.legend()
plt.grid()
plt.show()
```

### **An√°lisis de estabilidad**

Dado que los autovalores tienen una parte real de 0 y una parte imaginaria, el sistema es **marginalmente estable** y genera un movimiento oscilatorio sin crecimiento ni decaimiento exponencial. Esto significa que las trayectorias se mantienen en ciclos cerrados alrededor del origen, pero no se alejan ni se acercan, lo cual es caracter√≠stico de un **centro** en el plano fase.

#### R**esumen**
- El campo vectorial muestra el flujo en el plano, que revela el comportamiento oscilatorio.
- Las soluciones \( x(t) \) y \( y(t) \) oscilan con el tiempo, debido a los autovalores imaginarios.
- El sistema es **marginalmente estable** ya que las soluciones oscilan sin crecimiento ni amortiguamiento.

### **Punto 6.b**
> Resuelva los siguientes problemas de valores iniciales


$$
\mathbf{X}' = \begin{pmatrix}
2 & 4 \\
-1 & 6
\end{pmatrix} \mathbf{X}, \quad \mathbf{X}(0) = \begin{pmatrix}
-1 \\
6
\end{pmatrix}
$$

#### **Configurar e Inicializar las Variables y Matrices en SymPy**

Definimos la matriz $A$ del sistema y el vector de condiciones iniciales $X(0)$.

```python
import sympy as sp

# Definimos la variable de tiempo y el vector de inc√≥gnitas X(t)
t = sp.symbols('t')
x, y = sp.symbols('x y')
X = sp.Matrix([x, y])

# Matriz del sistema y vector de condiciones iniciales
A = sp.Matrix([[2, 4], [-1, 6]])
X0 = sp.Matrix([-1, 6])

# Mostramos la matriz A y la condici√≥n inicial
print("Matriz del sistema A:")
sp.pprint(A)

print("\nCondiciones iniciales X(0):")
sp.pprint(X0)
```

### **Calcular los Autovalores y Autovectores**

Ahora, calculamos los autovalores y autovectores de $A$, que son esenciales para resolver el sistema de ecuaciones diferenciales.

```python
# C√°lculo de autovalores y autovectores
autovalores = A.eigenvals()
autovectores = A.eigenvects()

# Verificamos el resultado de autovalores y autovectores
print("\nAutovalores:")
sp.pprint(autovalores)

print("\nAutovectores:")
for autovalor, multiplicidad, autovector in autovectores:
    print(f"Autovalor: {autovalor}, Multiplicidad: {multiplicidad}")
    print("Autovector asociado:")
    sp.pprint(autovector[0])
    print("\n")
```

### **Construir la Soluci√≥n General del Sistema**

Para construir la soluci√≥n general, usamos la combinaci√≥n lineal de los autovectores escalados por exponenciales de sus autovalores multiplicados por $t$.

```python
# Definimos las constantes de integraci√≥n
C1, C2 = sp.symbols('C1 C2')

# Creamos la soluci√≥n general con manejo para autovalores complejos
solucion_general = sp.Matrix([0, 0])
for i, (autovalor, multiplicidad, autovector) in enumerate(autovectores):
    # Combinamos autovectores con exponenciales de autovalores
    solucion_general += sp.symbols(f'C{i+1}') * autovector[0] * sp.exp(autovalor * t)

# Mostramos la soluci√≥n general
print("Soluci√≥n general del sistema:")
display(solucion_general)
```

### **Resolvemos las Constantes de Integraci√≥n usando la Condici√≥n Inicial**

Sustituimos $t = 0$ en la soluci√≥n general y usamos la condici√≥n inicial para resolver $C1$ y $C2$.

```python
# Igualamos la soluci√≥n en t = 0 con el vector de condiciones iniciales
ecuacion_inicial = sp.Eq(solucion_general.subs(t, 0), X0)

# Resolvemos para C1 y C2
constantes = sp.solve(ecuacion_inicial, (C1, C2))

# Soluci√≥n particular con las constantes espec√≠ficas
solucion_particular = solucion_general.subs(constantes)

# Mostramos la soluci√≥n particular
print("Soluci√≥n particular del sistema:")
display(solucion_particular)
```

\frac{d}{dt} \begin{pmatrix} x_1 \\ x_2 \end{pmatrix} = \begin{pmatrix} 1 & 8 \\ 1 & -1 \end{pmatrix} \begin{pmatrix} x_1 \\ x_2 \end{pmatrix} + \begin{pmatrix} e^{-t} \\ t e^t \end{pmatrix}
$$



```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Definimos la matriz A y el vector de t√©rminos independientes B(t)
A = sp.Matrix([[1, 8], [1, -1]])
B = sp.Matrix([sp.exp(-sp.Symbol('t')), sp.Symbol('t') * sp.exp(sp.Symbol('t'))])

# Paso 1: Resolver el sistema homog√©neo asociado (X' = A * X)
# Calculamos los valores y vectores propios de A
eigvals = A.eigenvals()
eigvecs = A.eigenvects()

print("Valores propios de A:", eigvals)
print("Vectores propios de A:", eigvecs)

# Paso 2: Construir la matriz fundamental
t = sp.symbols('t')
solutions_homogeneous = []
for eigen in eigvecs:
    lam = eigen[0]  # valor propio
    vec = eigen[2][0]  # vector propio correspondiente
    sol = sp.exp(lam * t) * vec  # soluci√≥n del sistema homog√©neo
    solutions_homogeneous.append(sol)

# Creamos la matriz fundamental Œ¶(t)
Phi = sp.Matrix.hstack(*solutions_homogeneous)
print("Matriz fundamental Phi(t):")
display(Phi)

# Paso 3: M√©todo de variaci√≥n de par√°metros
# Encontrar Phi_inv y Phi_inv * B
Phi_inv = Phi.inv()
u_prime = Phi_inv * B
u = sp.integrate(u_prime, t)  # Integrar para obtener u(t)
X_particular = Phi * u  # Soluci√≥n particular
print("Soluci√≥n particular Xp(t):")
display(X_particular)

# Paso 4: Soluci√≥n general
X_general = Phi * sp.Matrix(sp.symbols('c1 c2')) + X_particular
print("Soluci√≥n general X(t):")
display(X_general)

# Asignamos valores espec√≠ficos a las constantes c1 y c2
X_general_numeric = X_general.subs({'c1': 1, 'c2': 1})

# Para visualizar los resultados, convertimos las expresiones a funciones num√©ricas y graficamos.
t_vals = np.linspace(0, 5, 100)
x1_func = sp.lambdify(t, X_general_numeric[0], 'numpy')
x2_func = sp.lambdify(t, X_general_numeric[1], 'numpy')

x1_vals = x1_func(t_vals)
x2_vals = x2_func(t_vals)

plt.figure(figsize=(10, 5))
plt.plot(t_vals, x1_vals, label="x1(t)")
plt.plot(t_vals, x2_vals, label="x2(t)")
plt.xlabel("t")
plt.ylabel("Soluci√≥n")
plt.legend()
plt.title("Soluci√≥n del sistema no homog√©neo")
plt.grid()
plt.show()
```
